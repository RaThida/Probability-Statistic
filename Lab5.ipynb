```python
from num import Num
from db import Data1, data2
import matplotlib.pyplot as plt

num = Num()

x = [u[0] for u in Data1]
y = [u[1] for u in Data1]

print(f'x = {x}')
print(f'y = {y}')

ax1 = plt.subplot(1, 1, 1)
ax1.scatter(x, y, c='orange', marker='o', alpha=1)
plt.xlabel("Face width-to-height ratio")
plt.ylabel("Percent trusting")
plt.show()

zx = num.zscore(x)
zy = num.zscore(y)

print(f'Zx = {[f"{u:.3f}" for u in zx]}')
print(f'Zy = {[f"{u:.3f}" for u in zy]}')

sum_zx_zy = 0
for i in range(len(Data1)):
    sum_zx_zy += zx[i] * zy[i]

# correlation coefficient
r = sum_zx_zy / (len(Data1) - 1)
bold = '\033[1m'
print(f'\n{bold}r = {r:.3f}\n')
```

```python

import numpy as np
from num import Num
import matplotlib.pyplot as plt

x = [i[0] for i in data2]
y = [i[1] for i in data2]

print(f'x = {x}')
print(f'y = {y}')

# Scatter plot
ax = plt.subplot(1,1,1)
ax.scatter(x, y, c='r', marker='o', alpha=0.7, label='(x,y)')

# Draw regression line using numpy
a, b = np.polyfit(x, y, 1)
ax.plot(x, (a*np.array(x))+b, '-b', label='y = ax+b')

plt.xlabel("Days after injection")
plt.ylabel("Control average tumor size")
plt.legend()
plt.show()

n = len(data2)
sum_x = sum(x)
sum_y = sum(y)
sum_x2 = sum([i**2 for i in x])
sum_xy = sum([x[i] *y[i] for i in range(n)])
mean_x = sum_x / n
mean_y = sum_y / n
b1 = (n * sum_xy - sum_x * sum_y)
b2 = (n * sum_x2 - sum_x ** 2)
b = b1 / b2
a = mean_y - b * mean_x

print(f'a = {a}')
print(f'b = {b}')
print(f'y = {a} + {b}x')

# Calculate the tumor size if x = 35 days
print("Calculate the tumor size if x = 35 days")
y35 = a + b * 35
print(f'=> y = {a} + {b}*35')
print(f'=> y = {y35:.2f}')

bold = '\033[1m'
print(f'After 35 days, the tumor size would be: {bold}{y35:.2f} mm3')
```
